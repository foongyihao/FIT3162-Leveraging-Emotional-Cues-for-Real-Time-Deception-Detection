```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Path to precomputed HOF features
data_path = "../precomputed_hof_augmented"
video_name = "BM003_2NL"  # Change this to an actual video name in your dataset

# Load HOF features
hof_original = np.load(os.path.join(data_path, f"{video_name}_hof.npy"))
hof_aug1 = np.load(os.path.join(data_path, f"{video_name}_aug1_hof.npy"))
hof_aug2 = np.load(os.path.join(data_path, f"{video_name}_aug2_hof.npy"))

# Flatten for visualization
hof_original_flat = hof_original.reshape(hof_original.shape[0], -1)
hof_aug1_flat = hof_aug1.reshape(hof_aug1.shape[0], -1)
hof_aug2_flat = hof_aug2.reshape(hof_aug2.shape[0], -1)

# Combine for visualization
all_data = np.vstack([hof_original_flat, hof_aug1_flat, hof_aug2_flat])
labels = (["Original"] * len(hof_original)) + (["Aug1"] * len(hof_aug1)) + (["Aug2"] * len(hof_aug2))

# =======================
# 1 Histogram of Features
# =======================
plt.figure(figsize=(12, 5))
plt.hist(hof_original_flat.flatten(), bins=50, alpha=0.6, label="Original", color="blue")
plt.hist(hof_aug1_flat.flatten(), bins=50, alpha=0.6, label="Aug1", color="red")
plt.hist(hof_aug2_flat.flatten(), bins=50, alpha=0.6, label="Aug2", color="green")
plt.title("HOF Feature Distribution")
plt.xlabel("Feature Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# =======================
# 2 PCA Scatter Plot
# =======================
pca = PCA(n_components=2)
pca_result = pca.fit_transform(all_data)

plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:len(hof_original), 0], pca_result[:len(hof_original), 1], label="Original", alpha=0.6, color="blue")
plt.scatter(pca_result[len(hof_original):len(hof_original) + len(hof_aug1), 0],
            pca_result[len(hof_original):len(hof_original) + len(hof_aug1), 1], 
            label="Aug1", alpha=0.6, color="red")
plt.scatter(pca_result[len(hof_original) + len(hof_aug1):, 0],
            pca_result[len(hof_original) + len(hof_aug1):, 1], 
            label="Aug2", alpha=0.6, color="green")

plt.title("PCA Projection of HOF Features")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()

# =======================
# 3 t-SNE Visualization
# =======================
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(all_data)

plt.figure(figsize=(8, 6))
plt.scatter(tsne_result[:len(hof_original), 0], tsne_result[:len(hof_original), 1], label="Original", alpha=0.6, color="blue")
plt.scatter(tsne_result[len(hof_original):len(hof_original) + len(hof_aug1), 0], 
            tsne_result[len(hof_original):len(hof_original) + len(hof_aug1), 1], 
            label="Aug1", alpha=0.6, color="red")
plt.scatter(tsne_result[len(hof_original) + len(hof_aug1):, 0], 
            tsne_result[len(hof_original) + len(hof_aug1):, 1], 
            label="Aug2", alpha=0.6, color="green")

plt.title("t-SNE Projection of HOF Features")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Path to precomputed HOF features
data_path = "../precomputed_hof_augmented_4"
video_name = "BM003_2NL"  # Change this to an actual video name in your dataset

# Load HOF features
hof_original = np.load(os.path.join(data_path, f"{video_name}_hof.npy"))
hof_aug1 = np.load(os.path.join(data_path, f"{video_name}_aug1_hof.npy"))
hof_aug2 = np.load(os.path.join(data_path, f"{video_name}_aug2_hof.npy"))
hof_aug3 = np.load(os.path.join(data_path, f"{video_name}_aug3_hof.npy"))
hof_aug4 = np.load(os.path.join(data_path, f"{video_name}_aug4_hof.npy"))

# Flatten for visualization
hof_original_flat = hof_original.reshape(hof_original.shape[0], -1)
hof_aug1_flat = hof_aug1.reshape(hof_aug1.shape[0], -1)
hof_aug2_flat = hof_aug2.reshape(hof_aug2.shape[0], -1)
hof_aug3_flat = hof_aug3.reshape(hof_aug3.shape[0], -1)
hof_aug4_flat = hof_aug4.reshape(hof_aug4.shape[0], -1)

# Combine for visualization
all_data = np.vstack([hof_original_flat, hof_aug1_flat, hof_aug2_flat, hof_aug3_flat, hof_aug4_flat])
labels = (
    ["Original"] * len(hof_original) + 
    ["Aug1"] * len(hof_aug1) + 
    ["Aug2"] * len(hof_aug2) + 
    ["Aug3"] * len(hof_aug3) + 
    ["Aug4"] * len(hof_aug4)
)

# Colors for each set
colors = {
    "Original": "blue",
    "Aug1": "red",
    "Aug2": "green",
    "Aug3": "orange",
    "Aug4": "purple"
}

# =======================
# 1 Histogram of Features
# =======================
plt.figure(figsize=(12, 5))

for label, hof_data in zip(colors.keys(), [hof_original_flat, hof_aug1_flat, hof_aug2_flat, hof_aug3_flat, hof_aug4_flat]):
    plt.hist(hof_data.flatten(), bins=50, alpha=0.6, label=label, color=colors[label])

plt.title("HOF Feature Distribution")
plt.xlabel("Feature Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# =======================
# 2 PCA Scatter Plot
# =======================
pca = PCA(n_components=2)
pca_result = pca.fit_transform(all_data)

plt.figure(figsize=(8, 6))
start_idx = 0

for label, hof_data in zip(colors.keys(), [hof_original_flat, hof_aug1_flat, hof_aug2_flat, hof_aug3_flat, hof_aug4_flat]):
    end_idx = start_idx + len(hof_data)
    plt.scatter(pca_result[start_idx:end_idx, 0], pca_result[start_idx:end_idx, 1], 
                label=label, alpha=0.6, color=colors[label])
    start_idx = end_idx

plt.title("PCA Projection of HOF Features")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()

# =======================
# 3 t-SNE Visualization
# =======================
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(all_data)

plt.figure(figsize=(8, 6))
start_idx = 0

for label, hof_data in zip(colors.keys(), [hof_original_flat, hof_aug1_flat, hof_aug2_flat, hof_aug3_flat, hof_aug4_flat]):
    end_idx = start_idx + len(hof_data)
    plt.scatter(tsne_result[start_idx:end_idx, 0], tsne_result[start_idx:end_idx, 1], 
                label=label, alpha=0.6, color=colors[label])
    start_idx = end_idx

plt.title("t-SNE Projection of HOF Features")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()
plt.show()
```
